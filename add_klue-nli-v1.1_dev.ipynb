```python
import pandas as pd
```

```python
df = pd.read_json('./klue-nli-v1.1_dev.json')
```

```python
df
```

```python
df2 = df[['premise', 'hypothesis', 'gold_label']]
df2
```

```python
df2 = df2.rename(columns={'gold_label':'label'})
df2
```

```python
train = pd.read_csv("data/train_data.csv")
train = train[['premise', 'hypothesis', 'label']]
train
```

```python
new_train = pd.concat([train,df2])
new_train = new_train.reset_index(drop=True)
print(new_train)
```

```python
new_train.to_csv('./data/new_train_data.csv', index=False)
```

```python
snli_1_dev = pd.read_csv("./data/xnli.dev.ko.tsv", delimiter='\t', error_bad_lines=False)
snli_1_dev
```

```python
snli_1_dev = snli_1_dev.rename(columns={'sentence1':'premise', 'sentence2':'hypothesis', 'gold_label':'label'})
snli_1_dev
```

```python
new_train_all = pd.concat([new_train,snli_1_dev])
new_train_all = new_train_all.reset_index(drop=True)
print(new_train_all)
```

```python
new_train_all = new_train_all.drop_duplicates(['hypothesis'])
new_train_all = new_train_all.reset_index(drop=True)
print(new_train_all)
```

```python
new_train_all.isnull().sum()
```

```python
new_train_all = new_train_all.dropna(axis=0)
new_train_all.isnull().sum()
```

```python
new_train_all = new_train_all.reset_index(drop=True)
print(new_train_all)
```

```python
label_dict = {"entailment" : 0, "contradiction" : 1, "neutral" : 2}
```

```python
from tqdm import tqdm
for i, text in tqdm(enumerate(new_train_all.label)):
    new_train_all.label[i] = label_dict[text]
```

```python
print("premise 최대 길이:", new_train_all['premise'].map(len).max())
print("hypothesis 최대 길이:", new_train_all['hypothesis'].map(len).max())
```

```python
new_train_all.to_csv('./data/new_train_all.csv', index=False)
```
